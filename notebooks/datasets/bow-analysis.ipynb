```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
import pdb
import os
import re
from collections import defaultdict
from functools import reduce
```

```python
DATA_PATH = "../../data/raw/BoWFireDataset/"
TRAIN_PATH = DATA_PATH + "train/"
TEST_PATH = DATA_PATH + "dataset/img/"
```

# Helper Methods

```python
def describe_dataset(path: str, random: int = 5) -> None:
    images = os.listdir(path)
    print(f"Number of images: {len(images)}")

    # Find classes in the dataset
    regex = re.compile("([a-zA-Z_]+)([0-9]+)")
    classes = defaultdict(list)

    for image in images:
        data_class = regex.match(image).group(1)
        classes[data_class].append(image)
    
    print(f"Number of classes: {len(classes.keys())} - {list(classes.keys())}")

    # Plot bar chart of classes
    plt.figure(figsize=(10, 10))
    plt.title("Class distribution")
    plt.xlabel("Class")
    plt.ylabel("Number of images")
    plt.bar(list(classes.keys()), [len(v) for v in classes.values()])
    plt.show()
    
    images = os.listdir(path)
    # Randomly select images to display
    random_images = np.random.choice(images, random, replace=False)

    # Display images in subplots
    fig, ax = plt.subplots(1, 5, figsize=(10, 10))
    fig.suptitle("Randomly selected images from the dataset", y=0.7)
    fig.tight_layout()

    for i, image in enumerate(random_images):
        ax[i].imshow(plt.imread(path + image))
        ax[i].set_title(image)
        ax[i].axis('off')

    plt.show()
```

# Dataset Analysis

```python
describe_dataset(TRAIN_PATH)
```

```python
describe_dataset(TEST_PATH)
```
